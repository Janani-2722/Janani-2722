```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ipl=pd.read_csv("/content/matches.csv")
```

```python
#first five records
ipl.head()
```

```python
#rows and columns
ipl.shape
```

```python
#frequency of most man of the match awards
ipl["player_of_match"].value_counts()
```

```python
#top 10 players with most man of match awards
top_players=ipl['player_of_match'].value_counts()[0:10]
top_players
```

```python
top_players=ipl['player_of_match'].value_counts()[0:5]
top_players
```

```python
list(ipl['player_of_match'].value_counts()[0:5].key())
```

```python
#make barplot
plt.figure(figsize=(8, 5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()), list(ipl['player_of_match'].value_counts()[0:5]), color="purple")
plt.show()
```

```python
#frequency of result
ipl['result'].value_counts()
```

```python
#find no.of toss wins w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
#extracts the records where a team won battting first
batting_first=ipl[ipl['win_by_runs'] !=0]
batting_first.head()
```

```python
#make histogram
plt.figure(figsize=(8,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.show()
```

```python
#find no.of.wins w.r.t each team after batting first
batting_first['winner'].value_counts()
batting_first.head()
```

```python
#make barplot for top 3 teams with most after batting first
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]), color=['purple','yellow','blue'])

plt.show()
```

```python
#make piechart
plt.figure(figsize=(8,8))


plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extract those records where a team won after batting second
batting_second=ipl[ipl['win_by_wickets'] !=0]
batting_second.head()
```

```python
#make hist for frequency of wins w.r.t no.of.wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#find frequency of no.of.wins w.r.t each time after batting second
batting_second['winner'].value_counts()
batting_second.head()
```

```python
#make barplot for top 3 team with most wins after batting second
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=['blue','green','orange'])
plt.show()
```

```python
#make piechart
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
